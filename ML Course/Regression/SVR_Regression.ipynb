Support Vector Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Position_Salaries.csv")
X = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

```python
y = y.reshape(len(y),1)
print(y)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
X= sc_x.fit_transform(X)
y= sc_y.fit_transform(y)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel = 'rbf')
svr.fit(X,y) 
```

```python
sc_y.inverse_transform(svr.predict(sc_x.transform([[6.5]])).reshape(-1,1))
```

```python
plt.scatter(sc_x.inverse_transform(X),sc_y.inverse_transform(y),color = 'red',label='y')
plt.plot(sc_x.inverse_transform(X),sc_y.inverse_transform(svr.predict(X).reshape(-1,1)),label = 'Linear Model')
plt.title('SVR')
plt.xlabel('Position level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```
