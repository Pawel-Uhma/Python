```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Position_Salaries.csv")
X = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

```python
# from sklearn.model_selection import train_test_split
# X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 4)
X_pol = pf.fit_transform(X)
lr2 = LinearRegression()
lr2.fit(X_pol,y)
```

```python
plt.scatter(X,y,color = 'red',label='y')
plt.plot(X,lr.predict(X),label = 'Linear Model')
plt.title('LinearRegression')
plt.xlabel('Position level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

plt.scatter(X,y,color = 'red',label='y')
for n in [1,2, 5]:
    pf = PolynomialFeatures(degree = n)
    X_pol = pf.fit_transform(X)
    lr2 = LinearRegression()
    lr2.fit(X_pol,y)

    plt.plot(X,lr2.predict(X_pol),label = 'Polynomial level: n = %d'%n)
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
print("Linear: " + str(lr.predict([[6.5]])))
print("Polynomial: " + str(lr2.predict(pf.fit_transform([[6.5]]))))
```
